```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Visualizing anomalies

viz_cm = np.load("contact_maps/1A7EA_contact_map.npy")
viz_cp = np.load("contact_predictions/1A7EA_contact_pred.npy")

length = len(viz_cp)
errors = np.zeros((length, length))

for i in range(length):
    for j in range(length):
        if viz_cm[i][j] == 0 and viz_cp[i][j] > 0.8:
            errors[i][j] = 1

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,15))

axes[0].matshow(viz_cm)
axes[1].matshow(viz_cp)
axes[2].matshow(errors)
plt.show()

viz_cm = np.load("contact_maps/1A7EA_contact_map.npy")
viz_cp = np.load("contact_predictions/1A7EA_contact_pred.npy")

length = len(viz_cp)
errors = np.zeros((length, length))

for i in range(length):
    for j in range(length):
        if viz_cm[i][j] == 1 and viz_cp[i][j] < 0.2:
            errors[i][j] = 1

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,15))

axes[0].matshow(viz_cm)
axes[1].matshow(viz_cp)
axes[2].matshow(errors)
plt.show()
```
